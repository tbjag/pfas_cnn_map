```python
import rasterio
from rasterio.plot import show
import numpy as np
import matplotlib.pyplot as plt
```

```python
# https://www.sciencebase.gov/catalog/item/6604ffe7d34e64ff15495c1b
file_path1 = '../../data/USGS_1M_10_x37y531_WA_Olympic_Peninsula_Lidar_2017_B17.tif'
# https://www.sciencebase.gov/catalog/item/6604ffe0d34e64ff15495c13
file_path2 = '../../data/USGS_1M_10_x37y536_WA_Olympic_Peninsula_Lidar_2017_B17.tif'

file_path3 = '../../data/USGS_1M_10_x53y450_CA_CarrHirzDeltaFires_2019_B19.tif'
```

```python
def contrast_stretch(band):
    p2, p98 = np.percentile(band, (2, 98))
    return np.clip((band - p2) * 255 / (p98 - p2), 0, 255).astype(np.uint8)
```

```python
# Open the raster file
with rasterio.open(file_path1) as src:
    # Read the raster data
    raster_data = src.read(1)  # Read the first band
    
    # Apply contrast stretching
    stretched_data = contrast_stretch(raster_data)

    # Create a plot with enhanced resolution
    plt.figure(figsize=(10, 10), dpi=300)
    img = plt.imshow(stretched_data, cmap='gray', interpolation='nearest')
    plt.colorbar(img, label='Pixel values')
    plt.title("Contrast-Stretched Grayscale Raster Plot")
    plt.xlabel("Column Index")
    plt.ylabel("Row Index")
    plt.grid(False)  # Turn off the grid
    plt.show()
```

```python
# Open the raster file
with rasterio.open(file_path3) as src:
    # Read the raster data
    raster_data = src.read(1)  # Read the first band

    # Create a plot
    plt.figure(figsize=(10, 10))
    img = plt.imshow(raster_data, cmap='gray', interpolation='none')
    plt.colorbar(img, label='Pixel values')  # Attach the colorbar to the imshow object
    plt.title("Grayscale Raster Plot")
    plt.xlabel("Column Index")
    plt.ylabel("Row Index")
    plt.grid(False)  # Turn off the grid
    plt.show()
```
